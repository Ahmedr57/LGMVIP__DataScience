## Exploratory Data Analysis on Terrorism Dataset
#### Project Description
This project involves conducting an Exploratory Data Analysis (EDA) on a terrorism dataset to uncover key trends and insights. <br>
__The main objectives include:__
1. __Loading and Cleaning the Data:__ Understanding the structure of the dataset and handling missing values.
2. __Visualizing Trends Over Time:__ Analyzing the frequency of terrorist incidents over the years.
3. __Identifying Patterns:__ Exploring common attack types and the most affected regions and countries.
4. __Summarizing Insights:__ Drawing conclusions from the data to inform potential policy or security measures.
#### Key Findings and Insights
- __Regions Most Affected:__ The Middle East & North Africa has the highest frequency of terrorist incidents.
- __Trend Over Time:__ There has been a significant increase in terrorist activities starting from the 1970s.
- __Common Attack Types:__ Bombing/explosion is the most common type of terrorist attack.
- __Countries Most Affected:__ Iraq and Afghanistan are among the most severely impacted countries.

## Cleaning the dataset

```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
data.head()
```

```python
data=data[['Year','Month','Day','Country','State','Region','city','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
data.head()
```

## Visualizing the data

### Most attacks in terms of Terror Groups

```python
group_attacks = data.Group.value_counts().drop('Unknown').reset_index()[:11]
group_attacks.columns = ['Terrorist Group','Total Attacks']
group_attacks
```

#### Visual Representation

```python
group_attacks = data.Group.value_counts().drop('Unknown').reset_index()[:11]
group_attacks.columns = ['Terrorist Group','Total Attacks']
plt.subplots(figsize=(10,8))
sns.barplot(y=group_attacks['Terrorist Group'], x=group_attacks['Total Attacks'],palette='Spectral',edgecolor=sns.color_palette('dark',10))
plt.title('Number of total attacks by Terrorist Group')
plt.show()
```

### Terrorist activities by region in each year


```python
pd.crosstab(data.Year, data.Region).plot(kind='area',figsize=(15,7))
plt.title('Terrorist activities by region in each year')
plt.ylabel('Number of attacks')
plt.show()
```

### Number of total attacks by attack type

```python
attack_type = data.AttackType.value_counts().to_frame().reset_index()
attack_type.columns = ['Attack Type','Total Attacks']
plt.subplots(figsize=(15,7))
sns.barplot(x=attack_type['Attack Type'],y=attack_type['Total Attacks'],palette='viridis',edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=60)
plt.title('Number of total attacks by attack type')
plt.show()
```

### Top countries affected

```python
plt.subplots(figsize=(15,7))
sns.barplot(x=data['Country'].value_counts()[:15].index, y=data['Country'].value_counts()[:15].values, palette='mako')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()
```

### Most attacks per cities

```python
plt.subplots(figsize=(15,7))
sns.countplot(y='city',data=data,order=data.city.value_counts()[1:11].index,palette='mako')
plt.title('Attack per city')
```

### Types of target in terror attacks

```python
plt.subplots(figsize=(16,3))
sns.countplot(x="Target_type",data=data,order=data.Target_type.value_counts().index)
plt.xticks(rotation=60)
plt.title("Types of target")
plt.show()
```

### Total terrorist groups activities from 1970 to 2015

```python
groups_10 = data[data.Group.isin(data.Group.value_counts()[1:11].index)]
pd.crosstab(groups_10.Year, groups_10.Group).plot(color=sns.color_palette('Paired', 10))
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.xticks(range(1970, 2015, 5))
plt.ylabel('Total Attacks')
plt.title('Top Terrorist Groups Activities from 1970 to 2015')
plt.legend(labels=['Al-Shabaab',
                   'Boko Haraam',
                   'FMLN',
                   'IRA',
                   'ISIL',
                   'PKK',
                   'NPA',
                   'FARC',
                   'SL',
                   'Taliban'], loc='upper left')
```

### Analysis on customized data

```python
# Total Number of people killed in terror attack
killData = data.loc[:,'Killed']
print('Number of people killed by terror attack:', int(sum(killData.dropna())))
```

#### Attack Data

```python
attackData = data.loc[:,'AttackType']
typeKillData = pd.concat([attackData, killData], axis=1)
typeKillFormatData = typeKillData.pivot_table(columns='AttackType', values='Killed', aggfunc='sum')
typeKillFormatData
```

#### countyData

```python
countryData = data.loc[:,'Country']
countryKillData = pd.concat([countryData, killData], axis=1)
countryKillFormatData = countryKillData.pivot_table(columns='Country', values='Killed', aggfunc='sum')
countryKillFormatData
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size
```

#### 50 bar provides nice view

```python
labels = countryKillFormatData.columns.tolist()
labels = labels[:50] 
index = np.arange(len(labels))
transpoze = countryKillFormatData.T
values = transpoze.values.tolist()
values = values[:50]
```

#### convert float to int

```python
values = [int(i[0]) for i in values] 
```

#### color list for bar chart bar color 

```python
colors = ['red', 'green', 'gray', 'purple', 'yellow', 'magenta', 'black', 'blue', 'brown', 'orange'] 
```

#### For Chart Viusalization 

```python
fig, ax = plt.subplots(1, 1)
ax.yaxis.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size
plt.bar(index, values, color = colors, width = 0.9)
plt.ylabel('Killed People', fontsize=20)
plt.xlabel('Countries', fontsize = 20)
plt.xticks(index, labels, fontsize=18, rotation=90)
plt.title('Number of people killed by countries', fontsize = 20)
plt.show()
```
